# Lấy danh sách ảnh của 1 landmark

```python
lm_list = manager.get_landmark_list()

lm_list[:5]
```

```python
import matplotlib.pyplot as plt


def plot_images(images, titles, rows, cols):
    fig, axes = plt.subplots(rows, cols, figsize=(20, 40))
    for i, ax in enumerate(axes.flat):
        ax.imshow(images[i])
        ax.set_title(titles[i])
        ax.axis("off")
    plt.tight_layout()
    plt.show()
```

```python
images = manager.images_in_landmark(lm_list[1])[:10]
titles = [f"Image {i}" for i in range(len(images))]

plot_images(images, titles, 5, 2)
```
